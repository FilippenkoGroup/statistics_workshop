# Hypothesis Test
**Null hypothesis ($H_0$)**
: A testable hypothesis often stating that a feature of the **data is insigificant and due to random chance**. A hypothesis is testable if it's possible to create more samples out from the data.
    
**Alternative hypothesis ($H_A$)**
: Any other hypothesis often stating that a feature of the **data is significant and not due to random chance**.

## Population

A set of all possible object of interest from a group (e.g., the Milky Way stellar population contains all the starts in the Milky Way).

```python
die_values = (1, 2, 3, 4, 5, 6)
die_probs = [1/6] * 6

plt.bar(die_values, die_probs)

plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

## Sampling
To collect more data by taking a fixed number from the population or by simulating some hypothesis of what the population distribution.

```python
def roll(die_values, die_probs, nrolls=10):
    roll = np.random.choice(die_values, p=die_probs, size=nrolls)
    values, count = np.unique(roll, return_counts=True)
    return values, count

values, count = roll(die_values, die_probs, nrolls=10000)
plt.bar(values, count)

plt.grid()
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```

# Uncertainty

```python
dist = stats.norm(0, 1)
x = np.linspace(-5, 5, 100)

plt.plot(x, dist.pdf(x))
plt.hist(dist.rvs(10000), bins=20, density=True)
plt.show()
```
